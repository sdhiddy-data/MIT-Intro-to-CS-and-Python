### GUESS AND CHECK

- Process called **exhaustive enumeration**
- Applies to a problem where 
> you are able to **guess a value** for solution.<br>  you are able to check if the **solution is correct.**
- Keep guessing until you:
> You find a solution <br> Have guessed all the values.

```python
#Find the cube root of a perfect cube 
x = int(input('Enter an integer: '))# taking input from user
ans = 0 #creating an empty variable to store an answer
while ans**3 < abs(x): # we are checking what number when multiplied 3 times gives us x, and so if the cube exists it should be below the actual number.
    ans = ans + 1 #the loop ends once we find a number that whose ^3 is less than x
    
if ans**3 != abs(x): 
    print(x, 'is not a perfect cube') 
else: 
    if x < 0: 
        ans = -ans 
        print('Cube root of', x,'is', ans) 
```

Whenever you write a loop, you should think about an appropriate decrementing function.<br> This is a function that has the following properties: <br>  
It maps a set of program variables into an integer.<br> 
• When the loop is entered, its value is nonnegative.  
• When its value is ≤ 0, the loop terminates.  
• Its value is decreased every time through the loop.  
What is the decrementing function for the while loop above? It is 
abs(x) - ans**3. 

```python
x = int(input(" give a number: "))
ans = 0 
while ans**3 < abs(x):
    print("before:", ans)
    ans = ans + 1
    print("after:", ans)
```
